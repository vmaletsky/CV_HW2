```python
import cv2
import numpy as np
from scipy import ndimage
import matplotlib.pylab as plt
```

```python
img = cv2.imread('count/count1.jpg', cv2.IMREAD_COLOR)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
count=0
reth, thresh = cv2.threshold(img_gray,100, 255, cv2.THRESH_OTSU + cv2.THRESH_BINARY)
contours,h = cv2.findContours(thresh, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE )
for cnt in contours:
    if cv2.contourArea(cnt) > 50:
        count += 1
        cv2.drawContours(img, [cnt], 0, (255, 0, 0), 1)
print("Count: ", count)
plt.imshow(img)
```

```python
img = cv2.imread('count/count3.jpg', cv2.IMREAD_COLOR)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
kernel = np.ones((3,3),np.uint8)

reth, thresh = cv2.threshold(img_gray,100, 200, cv2.THRESH_OTSU + cv2.THRESH_BINARY)
black = np.zeros(img.shape).astype(img.dtype)  #black image to combine with
img_fg  = cv2.bitwise_or(img, black, mask = thresh)
imghsv = cv2.cvtColor(img_fg, cv2.COLOR_BGR2HSV)[:, :, 0]
thresh = cv2.erode(thresh, kernel, iterations=1)
contours,h = cv2.findContours(imghsv, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE )
count = 0
for cnt in contours:
    if cv2.contourArea(cnt) > 25:
        count += 1
        cv2.drawContours(img, [cnt], 0, (255, 0, 0), 1)
print("Count: ", count)
plt.imshow(img)
```

```python

img = cv2.imread('count/count11.jpg', cv2.IMREAD_COLOR)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
kernel = np.ones((3,3),np.uint8)

thresh = cv2.adaptiveThreshold(img_gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 9, 20)
thresh = cv2.dilate(thresh, kernel, iterations=1)
thresh = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)
black = np.zeros(img.shape).astype(img.dtype)  #black image to combine with
img_fg  = cv2.bitwise_or(img, black, mask = thresh)
imghsv = cv2.cvtColor(img_fg, cv2.COLOR_BGR2HSV)[:, :, 0]


contours,h = cv2.findContours(imghsv , cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE )
count = 0
for cnt in contours:
    if cv2.contourArea(cnt) > 15:
        count += 1
        cv2.drawContours(img, [cnt], 0, (255, 0, 0), 1)
print("Count: ", count)

plt.imshow(img)
```
